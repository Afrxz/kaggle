```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
inputs=df.drop(['CustomerType','CustomerGender','TravelPurpose'],axis='columns')
df.CustomerHappiness=df.CustomerHappiness.map({'Unhappy':0,'Happy':1})
df.head()
```

```python
inputs.ClassTravelled=inputs.ClassTravelled.map({'Economy':0,'PremiumEconomy':1,'Business':2})
inputs1=inputs.drop(['CustomerHappiness'],axis='columns')


inputs1.head()
```

```python
target=df.CustomerHappiness
target.head()
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(inputs1,target)
```

```python
tdf=pd.read_csv('test.csv')
tdf.head()
```

```python
tinputs=tdf.drop(['CustomerType','CustomerGender','TravelPurpose'],axis='columns')
tinputs.head()
```

```python
tinputs.ClassTravelled=tinputs.ClassTravelled.map({'Economy':0,'PremiumEconomy':1,'Business':2})



tinputs.head()
```

```python
sdf=model.predict(tinputs)
```

```python
abc=pd.DataFrame(sdf,columns=['CustomerHappiness'])
```

```python
abc.CustomerHappiness=abc.CustomerHappiness.map({0:'Unhappy',1:'Happy'})
abc.head()
```

```python
subdf=pd.read_csv('sample_submission.csv')
subdf.head()
```

```python
subdf1=subdf.drop(['CustomerHappiness'],axis='columns')
subdf1.head()
```

```python
finalsub=pd.concat([subdf1,abc],axis='columns')
finalsub
```

```python
finalsub.to_csv('kagglesubf.csv',index=False)
```
